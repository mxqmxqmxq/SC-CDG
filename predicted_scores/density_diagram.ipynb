```python
###compute the interaction with CDGs of each Net
output_file_path='/root/rlap-main_fe_gnn/predicted_scores/data/density/gene_inter_CDGs.txt'
with open(output_file_path, 'w') as file:
    for i in genes:
        list=[]
        for j in [GGNet,PPNet,PathNet]:
            a=j.query('name1=="{}" or name2=="{}"'.format(i,i))
            count_single_gene_Net=0
            for id,k in enumerate(c_drivers['name']):
                c=a.query('name1=="{}" or name2=="{}"'.format(k,k))
                count_single_gene_Net=count_single_gene_Net+len(c)
            list.append(count_single_gene_Net)   
        try:
            file.write(f"{i},{list[0]},{list[1]}, {list[2]}\n")
        except Exception as e:
            print(e)
file.close()
```
