```python
import networkx as nx

# 创建图
G = nx.Graph()
# 添加边
G.add_edge('A', 'B')
G.add_edge('B', 'C')
G.add_edge('A', 'C')

# 计算节点A和C之间的最短路径长度
length = nx.shortest_path_length(G, source='A', target='C')
print(length)  # 输出最短路径长度
```

```python
import networkx as nx

# Load the text file and construct the graph
with open('/root/autodl-tmp/HGDC-master/data/GGNet/GGNet.txt', 'r') as file:
    edges = [line.strip().split() for line in file.readlines()]

# Create a graph from the edges
G = nx.Graph()
G.add_edges_from(edges)

# Now let's demonstrate by computing the distance between two nodes
# Example nodes (can be changed to whatever the user specifies)
node1 = 'MACF1'
node2 = 'OR2T4'

# Calculate the shortest path
path_length = nx.shortest_path_length(G, source=node1, target=node2)
path_length
```
